## Exploratory Data Analysis
### Mobile Phone Price data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading the Slump test data
data= pd.read_csv('phone_data.csv')
```

```python
data.head(pd.set_option('display.max_columns',None))
```

```python
data.info()
```

```python
data.describe()
```

Insights from describe:
- More than 50% of the phones have RAM size between 1 to 3GB. The ranges from 64 MB to 12 GB.
- No. of cores in phone processor ranges from 1 to 10. At least 50% phones have less than 4 cores.
- While the highest internal storage is 512 GB, at least 50% phones have less than 16 GB storage.
- The minimum megapixels for rear and front cameras are 0, implying a small number of phones which do not have camera.
- Most phones have 2 SIM slots, though there are a small proportion that may have a single SIM or even 3 SIMs.

```python
data.describe(include=['O'])
```

Insights from describe (Objects):
- Android dominates among the 7 different classes of Operating Systems, with more than 95% of the phones.
- A majority of phones have modern-day essential features such as Bluetooth, Wi-Fi, GPS for navigation, 3G and 4G/LTE network capability, and Touchscreen.

## Univariate analysis

```python
# PLotting histograms give us the distribution for a singular feature.
sns.histplot(x='Battery capacity (mAh)',data=data)
```

```python
sns.histplot(x='Resolution x',data=data)
```

```python
sns.histplot(x='RAM (MB)',data=data)
```

```python
sns.distplot(data['Screen size (inches)'], hist=False) # shows the probability density function as a curve
```

```python
plt.figure(figsize=(12,5))
sns.distplot(data.Price, hist=False)  
```

Insights from Univariate analysis:
- Among most frequent battery capacities are 3000, 4000, an approx. 2600 milliampere-hour.
- A RAM size of 1000 MB is most frequent. For phones with a greater memory, the frequency of observations progressively decreases.
- Screen sizes of 5 to 5.5 inches are most common indicating their popularity among phone users.
- The most frequent pricing of phones is around 7,000. The average prices of phones is higher due to the few phone models that are highly priced.   

```python
# Countplot measures the count of labels for a categorical variable
sns.countplot(x='Touchscreen', data=data)
```

```python
# Using the hue parameter of Seaborn gives a backdrop to the data
sns.countplot(x='GPS', data=data, hue='Bluetooth')
```

- Majority of phones have Touchscreen, GPS, and Bluetooth.
- Among phones without GPS, a majority of models still have Bluetooth.

## Bivariate analysis

#### Relationship between categorical and numerical variables

```python
# Bar plot shows the measure of labels in a categorical variable with respect to a numerical variable
# Impact of Touchscreen on the phone price
sns.barplot(x='Touchscreen',y='Price',data=data)
```

```python
# DEfault values shown are for mean of Price
sns.barplot(x='3G',y='Price', data=data)
```

```python
# Multivariate analysis with bar plot when  the hue parameter is included
sns.barplot(x='4G/ LTE',y='Price', data=data, hue='3G')
```

- More pricing is seen for phones with Touchscreen feature available.
- Roughly equal mean price for phones with and without 3G
- Makes difference in price if 4G/LTE network is made available
- A good proportion of phones offering 3G also offer 4G/LTE network options.

```python
# Impact of processor cores on the phone price
plt.figure(figsize=(15,8))
sns.barplot(x='Processor',y='Price',data=data, hue='Operating system')
```

```python
# Which OS works out as more expensive or more reasonably priced. 
# Barplot gives the mean values of continuous variable as default
plt.figure(figsize=(10,5))
sns.barplot(x='Operating system',y='Price',data=data)
```

- Mean price of iOS phones the highest at around 45000, followed by Windows and Android phones.
- 6 processor cores followed by 4 cores lead to the highest average pricing of phones. 
- Only Android, Cyanogen and Windows operating systems offer 8 or more cores for the processor.
- iOS followed by Windows phones have the highest average prices followed by Android.

```python
# Using Catplot with categorical variables
```

```python
sns.catplot(x='Operating system',y='Price',data=data, aspect=2)
```

```python
#Catplot with swarm. Category with too many labels so swap the axes 
sns.catplot(y='Brand',x='Price',data=data, kind='swarm', height=15)
```

```python
# Relative movement of target with categorical variable. Hue for OS added herein. 
sns.catplot(y="Brand", x="Price", kind="point", data=data,height=20, hue="Operating system")
```

- Android is used as Operating system in the most expensive phones which are few in number.
- Since bulk of the android phones are in the affordable price range, iOS ends up with a higher average price, as do Windows phones.
- THe most expensive phones belong to Samsung brand. As seen earlier they use the Android operating system.

#### Relationship between numerical variables

```python
sns.lineplot(x='Resolution x',y='Resolution y',data=data)
```

```python
sns.lineplot(x=data['Screen size (inches)'],y=data['Resolution y'])
```

- Resolution x and y don't have a particularly strong linear relationship, but there is a trend visible.
- Resolution y and Screen size have a strong linear trend.

```python
# Different information to be gained from Scatterplots
```

```python
sns.scatterplot(x=data['Resolution x'],y=data['Resolution y'],data=data)
```

```python
sns.scatterplot(x=data['Screen size (inches)'],y=data['Resolution y'])
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x='Resolution y',y='Price', hue='Screen size (inches)', data=data)
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x=data['Rear camera'],y='RAM (MB)', data=data, hue='Front camera') 
```

```python
plt.figure(figsize=(15,10))
sns.relplot(x='Internal storage (GB)',y='Price',data=data,size='Screen size (inches)',aspect=2)
```

```python
plt.figure(figsize=(15,10))
sns.relplot(x='Battery capacity (mAh)',y='Price',data=data, col='4G/ LTE')
```

- Once resolution goes beyond 1500-2000 and screen size is greater or equal to 5 inches, the prices are upwards of 20,000.
- Better rear cameras correspond to memory >4000 MB RAM and also higher pixels for the front camera.
- It can be said that beyond 64GB, any increase in internal storage does not necessarily imply higher prices.
- For same battery capacity, the 4G/LTE equipped phones have more higher prices in general.
